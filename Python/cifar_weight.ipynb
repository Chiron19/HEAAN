```python
import torch
import os
model = torch.hub.load("chenyaofo/pytorch-cifar-models", "cifar10_resnet20", pretrained=True)

# print out the model's structure
print(model)
```

```python
# Print model's state_dict
print("Model's state_dict:")
for param_tensor in model.state_dict():
    print(param_tensor, "\t", model.state_dict()[param_tensor].size())
```

```python
def write_weight_to_txt_conv3x3(tensor, filename):
    with open(filename, 'w') as f:
        for i in range(tensor.size(0)):
            for j in range(tensor.size(1)):
                for k in range(tensor.size(2)):
                    for l in range(tensor.size(3)):
                        f.write(str(tensor[i][j][k][l].item()))
                        f.write(" ")
                f.write('\n')
    print(f"Done ---> {filename} ")

def write_weight_to_txt_conv1x1(tensor, filename):
    with open(filename, 'w') as f:
        for i in range(tensor.size(0)):
            for j in range(tensor.size(1)):
                f.write(str(tensor[i][j][0][0].item()))
                f.write(" ")
            f.write('\n')
    print(f"Done ---> {filename} ")

def write_weight_to_txt_bn(weight, running_var, bias, running_mean, filename):
    A = weight / running_var
    B = bias - running_mean * weight / running_var
    with open(filename, 'w') as f:
        for i in range(A.size(0)):
            f.write(str(A[i].item()))
            f.write(" ")
        f.write('\n')
        for i in range(B.size(0)):
            f.write(str(B[i].item()))
            f.write(" ")
        f.write('\n')
    print(f"Done ---> {filename} ")
    
def write_weight_to_txt_fc_weight(tensor, filename):
    with open(filename, 'w') as f:
        for i in range(tensor.size(0)):
            for j in range(tensor.size(1)):
                f.write(str(tensor[i][j].item()))
                f.write(" ")
            f.write('\n')
    print(f"Done ---> {filename} ")
def write_weight_to_txt_fc_bias(tensor, filename):
    with open(filename, 'w') as f:
        for i in range(tensor.size(0)):
            f.write(str(tensor[i].item()))
            f.write(" ")
        f.write('\n')
    print(f"Done ---> {filename} ")
```
